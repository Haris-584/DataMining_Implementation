## Cricket Dataset

```python
from scipy.io import arff
import pandas as pd

cricketdata = arff.loadarff("./data/CricketX_TRAIN.arff")
cricket = pd.DataFrame(cricketdata[0])
cricket.head()
```

```python
cricket.shape
```

```python
cricket.info()
```

```python
cricket.isna().any()
```

```python
cricket.isna().sum()
```

Missing Values

```python
import seaborn as sns
sns.heatmap(cricket.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```
