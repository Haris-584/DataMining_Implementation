##  Car  Dataset

```python
from scipy.io import arff
import pandas as pd

cardata = arff.loadarff("./data/Car_TRAIN.arff")
car = pd.DataFrame(cardata[0])
car.head()
```

```python
import seaborn as sns
sns.heatmap(car.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```
